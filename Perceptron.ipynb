```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
%matplotlib inline
```

```python
n_points = 500
np.random.seed(0)
top_region = np.array([np.random.normal(13,2,n_points),np.random.normal(12,2,n_points)]).T
bottom_region = np.array([np.random.normal(8,2,n_points),np.random.normal(6,2,n_points)]).T
X = np.vstack((top_region, bottom_region))
y = np.matrix(np.append(np.zeros(n_points), np.ones(n_points))).T
plt.scatter(X[:n_points,0], X[:n_points,1], color = 'b')
plt.scatter(X[n_points:,0], X[n_points:,1], color = 'r')
```

```python
model = Sequential()
model.add(Dense(units = 1, input_shape = (2,), activation = 'sigmoid'))
adam = Adam(lr = 0.1)
model.compile(adam, loss = 'binary_crossentropy', metrics = ['accuracy'])
h = model.fit(x=X, y=y, verbose=1, batch_size=50, epochs=500, shuffle='true')
```

```python
plt.plot(h.history['accuracy'])
plt.legend(['accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
```

```python
plt.plot(h.history['loss'])
plt.legend(['loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
```

```python
def plot_decision_boundary(X,y,model):
    x_span = np.linspace(min(X[:,0])-1, max(X[:,0])+1)
    y_span = np.linspace(min(X[:,1])-1, max(X[:,1])+1)
    xx,yy = np.meshgrid(x_span, y_span)
    xx_,yy_ = xx.ravel(), yy.ravel()
    grid = np.c_[xx_,yy_]
    predic_func = model.predict(grid)
    z = predic_func.reshape(xx.shape)
    plt.contourf(xx,yy,z)
    print(xx_)
    print(xx_.shape)
```

```python
plot_decision_boundary(X,y,model)
plt.scatter(X[:n_points,0],X[:n_points,1])
plt.scatter(X[n_points:,0],X[n_points:,1])
```
